```python
import scipy.io
import csv
import cv2
import matplotlib.pyplot as plt

# Path to the annotation.mat
path = 'PA-100K/annotation.mat'

# Load the .mat file
mat_data = scipy.io.loadmat(path)

# Extract the required data
attributes = [attr[0][0] for attr in mat_data['attributes']]
```

```python

test_image_names = [img[0][0] for img in mat_data['test_images_name']]
test_labels = mat_data['test_label']

# Choose an index (e.g., 0 for the first image)
index = 0

# Get image filename and its labels
image_name = test_image_names[index]
label_vector = test_labels[index]

# Map labels to attribute names
present_attributes = [attributes[i] for i, val in enumerate(label_vector) if val == 1]

# Print attributes
print(f"Image: {image_name}")
print("Attributes:", present_attributes)

# Display the image (make sure the path is correct)
img = cv2.imread(f"PA-100K/data/{image_name}")  # Update with your path
plt.imshow(img)
plt.title(f"Attributes: {', '.join(present_attributes)}")
plt.axis('off')
plt.show()
```
